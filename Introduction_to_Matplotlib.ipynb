```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.linspace(5, 100, 100)
y = np.linspace(5, 1000, 100)
plt.plot(x,y)
```

```python
plt.plot(x,y)

plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("Ohm's law")

plt.xlim([20,80])
plt.ylim([200,800])
plt.show()
```

```python
x=np.linspace(0,10,20)
y=x*2
plt.plot(x,y, 'b+')
plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("Ohm's Law")
plt.show()
```

```python
x=np.linspace(0,5,10)
y=np.linspace(3,6,10)

plt.plot(x, y, 'r-', x, y**2, 'b+', x, y**3, 'g^')
plt.show()

```

```python
x=np.linspace(1,10,100)
y=np.log(x)

plt.figure(1)

plt.subplot(121)
plt.title("y=log(x)")
plt.plot(x,y)

plt.subplot(122)
plt.title("y=log(x)**2")
plt.plot(x,y**2)
plt.show()
```

```python
x=np.linspace(1,10,100)
plt.figure(2)
plt.subplot(2,2,1)
plt.title("Linear")
plt.plot(x,x)
plt.subplot(2,2,2)
plt.title("cubic")
plt.plot(x,x**3)
plt.figure(3)
plt.subplot(2,2,1)
plt.title("log")
plt.plot(x,np.log(x))
plt.subplot(2,2,2)
plt.title("exponential")
plt.plot(x,x**2)
plt.show()
```

```python
x=np.linspace(5,9,5000)
y=np.linspace(2,4,8)
plt.figure(1)
plt.subplot(4,4,1)
plt.title("linear")
plt.plot(y,y)

plt.subplot(4,4,2)
plt.title("square")
plt.plot(x,x**2)

plt.subplot(4,4,3)
plt.title("cube")
plt.plot(x,x**3)

plt.subplot(4,4,4)
plt.title("log")
plt.plot(x,np.log(x))

plt.figure(2)
plt.subplot(4,4,1)
plt.title("linear")
plt.plot(y,y**4)

plt.subplot(4,4,2)
plt.title("square")
plt.plot(x,x**2)

plt.subplot(4,4,3)
plt.title("cube")
plt.plot(x,x**3)

plt.subplot(4,4,4)
plt.title("log")
plt.plot(x,np.log(x))

plt.figure(3)
plt.subplot(4,4,1)
plt.title("linear")
plt.plot(y,y**2)

plt.subplot(4,4,2)
plt.title("square")
plt.plot(x,x**2)

plt.subplot(4,4,3)
plt.title("cube")
plt.plot(x,x**3)

plt.subplot(4,4,4)
plt.title("log")
plt.plot(x,np.log(x))
```
